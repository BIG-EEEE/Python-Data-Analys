## Import needed library

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
da=pd.read_csv("Urban Air Quality and Health Impact Dataset.csv")
pd.set_option("display.max_rows",None)
da
```

```python
#check for null values
da.isnull().any()

# checking and correcting data types
da.dtypes
da["Month"]= da["Month"].astype("object")
```

```python
#rename rows
da["Month"]=da.loc[da['Month'] == "0:999"] = 'September'
da["Month"]
```

```python
#validating changes
da.head()
```

```python
# dropping unwanted columns
da=da.drop(["datetimeEpoch","stations"],axis=1)
```

```python
print(da.columns)
```

```python
#replace NaN values
da=da.fillna(0)
```

```python
#checking null value existence
da.isnull().any()
```

```python
da.head(6)
```

## Answering questions

```python
#Question 1: what is the average temperature?
Avg_temp= da["temp"].mean()
Avg_temp
```

```python
#Question 2: Which date recorded the highest and lowest humidity value
highest_hmdty_date = da.loc[da['humidity'].idxmax(), 'datetime']
lowest_hmdty_date = da.loc[da['humidity'].idxmin(), 'datetime']

# Output the results
print(f"date with highest humidity value record: {highest_hmdty_date}")
print(f"date with lowest humidity value record: {lowest_hmdty_date}")
```

```python
#Question 3: which city recorded the highest and lowest precipitation and temperature values
highest_precip_city = da.loc[da['precip'].idxmax(), 'City']
lowest_precip_city = da.loc[da['precip'].idxmin(), 'City']

# Find the city with the highest and lowest temperature
highest_temp_city = da.loc[da['temp'].idxmax(), 'City']
lowest_temp_city = da.loc[da['temp'].idxmin(), 'City']

# Output the results
print(f"City with highest precipitation: {highest_precip_city}")
print(f"City with lowest precipitation: {lowest_precip_city}")
print(f"City with highest temperature: {highest_temp_city}")
print(f"City with lowest temperature: {lowest_temp_city}")
```

```python
# Question 4: Temperature and humidity are inversely proportional, how true is that?. Use correlation and graphical representation to validate.
tmp=da["temp"]
hmdty=da["humidity"]
#calculating correlation
corr=np.corrcoef(tmp.values.flatten(),hmdty.values.flatten())[0,1]
print(f"Correlation value between temperature and humidity: {corr}")

#Graphical representation
fig, ax = plt.subplots(1, 3, figsize=(22,6))
# Plot temperature on ax[0]
ax[0].plot(tmp.head(30), label="Temperature", color="red")
ax[0].fill_between(tmp.head(30).index, tmp.head(30), color="red", alpha=0.05)
ax[0].set_title('Temperature Trend',fontweight="bold")
ax[0].set_xlabel("Counts",fontweight="bold")
ax[0].set_ylabel("Temperature",fontweight="bold")
ax[0].grid()
ax[0].legend()
# Plot humidity on ax[1]
ax[1].plot(hmdty.head(30), label="Humidity", color="blue")
ax[1].fill_between(hmdty.head(30).index, hmdty.head(30), color="blue", alpha=0.05)
ax[1].grid(linestyle="--")
ax[1].set_title('Humidity Trend',fontweight="bold")
ax[1].set_xlabel("count",fontweight="bold")
ax[1].set_ylabel("Humidity values",fontweight="bold")
ax[1].legend()
# Create a twin axis for ax[2]
ax2 = ax[2].twinx()
# Plot temperature on ax[2] and humidity on ax2
ax[2].plot(tmp.head(30), label="Temperature", color="red")
ax2.plot(hmdty.head(30), label="Humidity", color="blue")
#ax2.grid()
ax[2].grid()
ax[2].set_xlabel("count")
ax[2].set_ylabel("Temperature",color="red")
ax2.set_ylabel("humidity",color="blue")
ax[2].set_title('Correlation Between Temperature vs Humidity',fontweight="bold",fontsize=16)
ax[2].legend(loc='upper left')
ax2.legend(loc='upper right')
plt.tight_layout()
plt.show()
fig.savefig("plot.png",dpi=350)
```

```python
# Question 5: Find average dewpoint temperature of the individual cities
avg_dp=da.groupby("City").agg(mean_dewpoint=("dew","mean"))
avg_dp
```

```python
# Question 6: find the relationship between temperature and Health_Risk_score
x = [str(i) for i in range(1, 21)]
fig, ax1=plt.subplots(figsize=(14,5))
ax1.plot(x,da["temp"].head(20),color="red")
ax2=ax1.twinx()
ax2.plot(x, da["Health_Risk_Score"].head(20),color="green")
ax1.grid()
```
